### Step 4: Handle Missing Values.

```python
dataset = pd.read_csv(data_url)

# First, identify missing values using isnull() function.
missingValues = dataset.isnull().sum()
print("Identifying Missing Values...")
print(missingValues)
```

```python
# Then, analyze patterns of missing values by computing missing value percentages.
missingFeatures = dataset.isnull().mean()
print("\nMissing Values by Feature:")
print(missingFeatures)
print('\nMissing values are concentrated in "Age", "Cabin", and "Embarked" features.')
```

```python
# Lastly, perform deletion by dropping columns with high missing value percentages.
dataset = dataset.dropna(subset=["Age"])
dataset = dataset.drop(columns=["Cabin"])
dataset = dataset.dropna(subset=["Embarked"])

# Verify that deletion was performed on the three columns by calling the info() function.
dataset.info()
```

### Step 5: Clean noisy data.

```python
# First, identify outliers using boxplots
featuresWithOutliers = ["Age", "SibSp", "Parch", "Fare"]
dataset[featuresWithOutliers].boxplot(figsize=(10, 10))
plt.title('Boxplot of Features with Outliers')
plt.show()
```

```python
# Apply the capping technique to handle outliers by replacing outliers 
# with the nearest non-outlier value.
for feature in featuresWithOutliers:
    # NOTE: Q1 = 0.25 and Q3 = 0.75 
    percentiles = dataset[feature].quantile([.25, .75]).values
    dataset[feature] = np.clip(dataset[feature], percentiles[0], percentiles[1])

# Verify that capping was performed correctly.
dataset[featuresWithOutliers].boxplot(figsize=(10, 10))
plt.title('Boxplot of Features with Outliers')
plt.show()
```

### Step 6: Identify Redundant Features.

```python
# Calculate correlations between features by removing highly correlated features
# in order to avoid multicollinearity.
# NOTE: we are dropping Passenger ID, Name, and Ticket from the new dataset 
# since they're all unique.
newDataset = dataset.drop(columns=["PassengerId", "Name", "Ticket"])
```

```python
print("Calculating correlations between features...")
newDataset.corr()
```

No multicollinearity detected.

### Step 7: Address Inconsistencies.

Inconsistencies were identified during Step 5, data formats are standardized, and consistent labeling for categorical features are ensured.

### Step 8: Normalization.

```python
from sklearn import preprocessing

# Identify numerical features by using the select_dtypes() methods to select numerical columns.
numericalCols = newDataset.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Normalize values to have unit length using normalize from scikit-learn.
for col in numericalCols:
    dataset[col] = preprocessing.normalize([dataset[col]])[0]
```

```python
# Display the normalized dataset.
print("Normalized dataset:")
dataset
```

### Step 9: Categorical Feature Encoding.

```python
# Convert categorical features into binary dummy variables using get_dummies() from pandas.
preprocessedData = pd.get_dummies(dataset, columns=["Sex", "Embarked"])
```
